### **1. Forecasting Problem**

Collaborating as a team participating in Kaggle’s BUSA Predictive Analytics Competition, we are tasked with forecasting vehicle prices as accuratey as possible, based on various features provided in the dataset.  Accurate price forecasting offers significant value to several key stakeholders:

- **Insurance Providers**: Accurate price forecasts allow insurers to assess vehicle depreciation and set precise premiums, aligning coverage with the vehicle's true value and reducing financial risk.

- **Consumers**: Price forecasts guide buyers on when to purchase or negotiate better deals, leading to smarter decisions and potential savings.

### **2. Evaluation Criteria**

The accuracy/performance of the forecasting models is evaluated using the Mean Absolute Percentage Error (MAPE). The formula  is as follows:

<center>
$MAPE = \frac{1}{n} \sum_{i=1}^n \left| \frac{y_i - \hat{y}_i}{y_i} \right| \times 100$
</center>

Where:

- $n$: the number of observations
- $y_i$: the actual car price in the dataset
- $\hat{y}_i$: the predicted car price from the model

MAPE quantifies the average percentage error between predicted and actual values, with lower scores indicating higher model accuracy. The competition leaderboard ranks models based on MAPE, rewarding the lowest score for the best forecast.

MAPE is a straightforward and easy-to-interpret metric that provides a balanced view of model performance in datasets with potential outliers. Unlike RMSE and MSE, MAPE penalizes larger errors more than smaller ones, encouraging accurate predictions for lower-priced cars. Its focus on relative errors effectively contextualizes performance in datasets with high variability.

### CLEAN NUMERICAL VARIABLES

```python
## Task 2, Question 1 Code Here
# Define function for cleaning all numerical values
def preprocess_numerical_features(dat):
    # Define columns that may contain 'in' and need unit removal
    columns_with_in = ['back_legroom', 'front_legroom', 'height', 'wheelbase', 'width']

    # Remove 'in' from relevant columns if they exist, handle '--' values
    for col in columns_with_in:
        if col in dat.columns:
            # Replace '--' and other non-numeric values with NaN
            dat[col] = dat[col].replace(['--', ''], None)
            dat[col] = pd.to_numeric(dat[col].str.replace(' in', ''), errors='coerce')

    # Handle maximum_seating column to remove 'seats' and convert to integer
    dat['maximum_seating'] = pd.to_numeric(dat['maximum_seating'].str.replace(' seats', ''), errors='coerce', downcast = 'integer')

    return dat

# Apply the defined function to train and test dataset
train_dat = preprocess_numerical_features(train_dat)
test_dat = preprocess_numerical_features(test_dat)
```

**First, both the train and test datasets are imported using the pandas function `read_csv` to load the data for further processing.**

**Secondly, I define the function to preprocess the numerical features, taking the dataset as the parameter, which contains 3 steps:**
1. I defined a list of columns (`back_legroom, front_legroom, height, wheelbase, width`) where the values might include units like "in" (inches) or non-numeric characters like "--".

2. Using a for loop, I iterated through these columns and applied a series of transformations:
- I replaced non-numeric values (like "--") with NaN.
- I used str.replace() to remove the "in" unit from these columns.
- I converted the cleaned strings to numeric values with pd.to_numeric, using errors='coerce' to convert invalid values to NaN.

3. For the `maximum_seating` column, I used str.replace() to remove the word "seats" and converted the result to an integer using pd.to_numeric.

### FEATURE ENGINEERING

**Task 2, Question 2** Create at least 5 new features from the existing numerical variables which contain multiple items of information, for example you could extract maximum torque and torque rpm from the torque variable.  
(2 marks)

```python
## Task 2, Question 2 Code Here

# Define function for feature engineering
def feature_engineering(dat):
    # Extract maximum horsepower and RPM from 'power' column
    dat['max_power'] = pd.to_numeric(dat['power'].str.extract(r'(\d+) hp')[0], errors='coerce')  # Extract maximum horsepower from power
    dat['power_rpm'] = pd.to_numeric(dat['power'].str.extract(r'@ (\d{1,3},\d{3}|\d+) RPM')[0].str.replace(',', ''), errors='coerce') #Extract rpm from power

    # Extract maximum torque and RPM from 'torque' column
    dat['max_torque'] = pd.to_numeric(dat['torque'].str.extract(r'(\d+) lb-ft')[0], errors='coerce') # Extract maximum torque value from torque
    dat['torque_rpm'] = pd.to_numeric(dat['torque'].str.extract(r'@ (\d{1,3},\d{3}|\d+) RPM')[0].str.replace(',', ''), errors='coerce') # Extract rpm from torque

    # Extract number of cylinders and cylinder arrangements from 'engine_type' column
    dat['num_cylinders'] = pd.to_numeric(dat['engine_type'].str.extract(r'(\d+)')[0], errors='coerce', downcast='integer')
    dat['cyl_arrangement'] = dat['engine_type'].str[0]

    # Calculate the age of the vehicles using datetime library
    dat['vehicle_age'] = dt.datetime.now().year - dat['year']

    # Extract listed year, month, date from 'listed_date' columns
    dat['listed_date'] = pd.to_datetime(dat['listed_date'], errors='coerce')
    dat['listed_year'] = dat['listed_date'].dt.year
    dat['listed_month'] = dat['listed_date'].dt.month
    dat['listed_day'] = dat['listed_date'].dt.day

    #drop old columns that contains the same information
    dat.drop(['engine_type','power','torque','year','listed_date'], axis=1, inplace = True)

    return dat

    #Think more whether should compute some ratios, etc. or could be done in step 5

# Apply the defined function to train and test dataset
train_dat = feature_engineering(train_dat)
test_dat = feature_engineering(test_dat)
```

In this step, we define a function to perform feature engineering, which involves transforming and extracting useful information from the dataset columns. This process focuses on key variables such as power, torque, engine_type, and year to create additional features that enhance the dataset's usefulness for further analysis and modeling.

**1. Extract maximum horsepower and relevant RPM from the `power` column:**
- Maximum horsepower: We used a regular expression (regex) to extract the numeric value before the "hp" unit in the power column, which represents the vehicle's maximum horsepower. After extracting, the value is converted to a numeric type to ensure compatibility with further calculations and analysis.
- Power RPM: The RPM value, which typically follows the format "a,bcd" in the power column, was extracted using regex and commas were removed to standardise the data. Like horsepower, the RPM value is then converted to a numeric type.
- Both horsepower and RPM are important indicators of a vehicle's performance, especially for evaluating its acceleration capabilities and engine power.

**2. Extract maximum torque and relevant RPM from the `torque` column:**
- Maximum torque: We applied a similar extraction process as used for horsepower. Using regex, we extracted the numeric value preceding "lb-ft" in the torque column.
- Torque RPM: Like the power RPM, the RPM value for torque was extracted using the same regex logic and converted to a numeric format.
- Since torque is a key measure of vehicle power at lower speeds, separating these values allows for a more detailed analysis of engine performance.

### MISSING VALUES IMPUTATION

**Task 2, Question 3**: Impute the missing values for all features in both the training and test datasets.   
(2 marks)

```python
## Task 2, Question 3 Code Here

# Define function for imputing missing values
def impute_missing_values (dat):
    ## Categorical variables

    # Define columns where missing values can be imputed with mode, grouped by make_name and model_name
    cat_cols_group_by_make_model = ['exterior_color','interior_color', 'transmission', 'transmission_display', 'wheel_system', 'cyl_arrangement', 'fuel_type']

    # Function to safely compute the mode within a group and fallback to the global mode if no mode is available
    def compute_mode(series, global_mode):
        if not series.mode().empty:
            return series.mode()[0]
        return global_mode  # Fallback to the global mode if no mode in the group

    # Impute missing values for the specified columns, grouped by make_name and model_name
    for col in cat_cols_group_by_make_model:
        # Compute the global mode of the column
        global_mode_col = dat[col].mode()[0]
        # Apply the group-based imputation with fallback to the global mode
        dat[col] = dat.groupby(['make_name', 'model_name'])[col].transform(lambda x: x.fillna(compute_mode(x, global_mode_col)))

    # Impute missing values for franchise_dealer with mode, grouped by dealer_zip
    global_mode_franchise = dat['franchise_dealer'].mode()[0]  # Compute global mode for franchise_dealer
    dat['franchise_dealer'] = dat.groupby('dealer_zip')['franchise_dealer'].transform(lambda x: x.fillna(compute_mode(x, global_mode_franchise)))

    ## Numerical variables

    # Define columns of which missing values can be filled with mean of those grouped by body_type
    cols_group_by_body_type = ['back_legroom','front_legroom', 'width', 'height', 'maximum_seating', 'wheelbase']

    # Impute missing values for above columns with means of those grouped by body_type
    for col in cols_group_by_body_type:
        dat[col] = dat.groupby('body_type')[col].transform(lambda x: x.fillna(x.mean()))

    # Define columns where missing values can be imputed with mean, grouped by make_name and model_name
    num_cols_group_by_make_model = ['city_fuel_economy', 'engine_displacement', 'max_power', 'power_rpm', 'highway_fuel_economy', 'max_torque', 'torque_rpm', 'num_cylinders']

    # Function to safely compute the mean within a group and fallback to the global mean if no mean is available
    def compute_mean(series, global_mean):
        if series.notna().sum() == 0:  # Check if all values in the group are NaN
            return global_mean  # Fallback to the global mean if no mean in the group
        return series.mean()  # Otherwise, return the group mean

    # Impute missing values for the specified columns, grouped by make_name and model_name
    for col in num_cols_group_by_make_model:
        # Compute the global mean of the column
        global_mean = dat[col].mean()
        # Apply the group-based imputation with fallback to the global mean
        dat[col] = dat.groupby(['make_name', 'model_name'])[col].transform(lambda x: x.fillna(compute_mean(x, global_mean)))

    # Impute missing values for latitude with means of those grouped by city
    dat['latitude'] = dat.groupby('city')['latitude'].transform(lambda x: x.fillna(x.mean()))
    dat['latitude'] = dat['latitude'].transform(lambda x: x.fillna(x.mean())) #fill with global mean value in case there are still missing values

    # Median imputation for seller_rating as the data is left-skewed
    dat['seller_rating'] = dat.groupby(['make_name', 'model_name', 'franchise_dealer'])['seller_rating'].transform(lambda x: x.fillna(x.median()))

    # Utilise iterative imputer to impute missing values for 'mileage' and 'saving_amount'
    # Select impactful features for MICE imputation
    impactful_features = ['vehicle_age', 'seller_rating', 'daysonmarket', 'horsepower', 'savings_amount', 'mileage']
    impute_data = dat[impactful_features]

    # Apply iterative imputer to impute missing values
    imp_imputer = IterativeImputer(max_iter=10, random_state=0, min_value=0)

    # Fit and transform the data to fill missing values
    imputed_data = imp_imputer.fit_transform(impute_data)

    # Replace the imputed values back into the original dataset
    dat[impactful_features] = imputed_data

    return dat


# Apply the defined function to train and test dataset
train_dat = impute_missing_values(train_dat)
test_dat = impute_missing_values(test_dat)
```

In this part, I created a function `impute_missing_values`, which is designed to handle missing values in both categorical and numerical columns. The function will use different imputation techniques, including mode, mean, and iterative imputation, depending on the type of data and the context of the missing values. Below is a step-by-step outline of what each part of the function does and the reasoning behind it.

**1. Imputing missing values for categorical variables:**
- I first defined a list of categorical columns where missing values should be imputed using the mode within similar groups (based on `make_name` and `model_name`). This is to ensure that the imputation is contextually relevant to the specific type of vehicle.
- Then, a helper function `compute_mode` is also created, which calculates the mode within each group. If no mode exists for a group (e.g., if all values are missing), it falls back to the global mode, making sure there is no missing values after imputation.
- For the  `franchise_dealer` column, we used a similar process, grouping by `dealer_zip` and imputing the missing values using the mode of the group or the global mode.

**2. Imputing missing values for numerical variables:**
- I also defined a list of numerical columns  that can be grouped by `body_type` (`back_legroom`,`front_legroom`, `width`, `height`, `maximum_seating`, `wheelbase`) and using the group mean for imputation. This is because vehicles with the same `body_type` (e.g., SUVs, sedans) tend to have similar dimensions and seating configurations.
- For other numerical columns like `engine_displacement`, `city_fuel_economy`, etc., we imputed missing values based on `make_name` and `model_name` groups, using the group mean. If no group mean was available, we fell back to the global mean. By doing so, the imputed values are consistent with the vehicle’s specifications, which likely vary by make and model.
- For `latitude`, I applied the same logic by grouping by `city` to impute missing values based on the group mean. If any values were still missing after group imputation, I filled them with the global mean.
- For `seller_rating`, I used median imputation instead of mean, as this column is left-skewed and median is robust to outliers. I grouped it by `make_name, model_name, and franchise_dealer` to fill in missing values.

**3. Imputing missing values using iterative imputation (MICE):**
- For columns like `mileage` and `savings_amount`, I applied iterative imputation (MICE) to capture complex relationships between variables and produces more accurate and informed imputations for these 2 important features. This method uses regression models in multiple iterations to impute missing values based on other relevant features (e.g., vehicle_age, seller_rating, etc.).

**4. Apply the function to train and test datasets:**
- Finally, we applied the impute_missing_values function to both the `train` and `test` datasets, ensuring both datasets are consistently processed and free of missing values.

```python
# Step 1: Define a function to calculate top 5 values for each categorical column from both datasets combined
def get_top_5_values(train_dat, test_dat):
    top_5_dict = {}
    combined_dat = pd.concat([train_dat, test_dat], axis=0)  # Combine both datasets

    # Get categorical columns, excluding 'vin'
    cat_cols = combined_dat.select_dtypes(include='object').columns.drop('vin', errors='ignore')

    for col in cat_cols:
        # Get top 5 values for each column based on combined data
        top_5_values = combined_dat[col].value_counts().nlargest(5).index
        top_5_dict[col] = top_5_values

    return top_5_dict

# Step 2: Define function for encoding categorical variables consistently across train and test data
def encode_categorical_vars(dat, top_5_dict):
    # Select categorical columns
    cat_cols = top_5_dict.keys()

    for col in cat_cols:
        # Step 1: Replace values not in the top 5 with 'Other'
        top_5_values = top_5_dict[col]
        dat[col] = dat[col].apply(lambda x: x if x in top_5_values else 'Other')

        # Step 2: Create dummy variables
        dummies = pd.get_dummies(dat[col], prefix=col, drop_first=True)

        # Step 3: Concatenate dummies to original dataframe
        dat = pd.concat([dat, dummies], axis=1)

        # Step 4: Drop the original column
        dat.drop(columns=[col], inplace=True)

    return dat

# Step 3: Apply encoding and align columns for consistency
top_5_dict = get_top_5_values(train_dat, test_dat)

# Encode both train and test datasets
train_dat = encode_categorical_vars(train_dat, top_5_dict)
test_dat = encode_categorical_vars(test_dat, top_5_dict)

# Step 4: Align columns in test_dat with train_dat
missing_cols = set(train_dat.columns) - set(test_dat.columns)
for col in missing_cols:
    test_dat[col] = 0  # Add missing columns with 0 values in test set

# Ensure column order consistency
test_dat = test_dat[train_dat.columns]
```

**1. Identify categorical columns for encoding:**

**2. Define a function to encode the top 5 most frequent values and create dummy variables:**
- For each categorical column, I identified the top 5 most frequent values using value_counts() and extracted these into a list. For values in the column that were not among the top 5, I replaced them with "Other". Furthermore, the multiple-value observations where values in the column were formatted as "a/b" are also replaced with "Other". This helps reduce the complexity of the dataset and ensures that the most important categories are retained.
- I then created dummy variables (one-hot encoding) for the categorical column, using pd.get_dummies(). The drop_first=True argument avoids data redundancy by dropping one of the categories.

```python
## Task 2, Question 5 Code Here

# Remove outliers for all numerical variables (capping at 1st and 99th percentiles)
def remove_outliers(dat):
    numeric_cols = dat.select_dtypes(include='number').columns  # Select all numerical columns
    for col in numeric_cols:
        lower_bound = dat[col].quantile(0.01)
        upper_bound = dat[col].quantile(0.99)
        dat[col] = dat[col].clip(lower_bound, upper_bound)  # Cap values at 1st and 99th percentiles
    return dat

# Apply all transformations to both train and test datasets
def add_preprocess_data(dat):
    # Log transform skewed variables ('mileage')
    dat['log_mileage'] = np.log1p(dat['mileage'])

    # Possible feature creation
    dat['mileage_per_year'] = dat['mileage'] / dat['vehicle_age'].replace(0, 1)  # Replace 0 age to avoid division by zero
    dat['hp_weight_ratio'] = dat['horsepower'] / dat['maximum_seating']
    dat['fuel_economy_diff'] = dat['highway_fuel_economy'] - dat['city_fuel_economy']
    dat['displacement_per_cylinder'] = dat['engine_displacement'] / dat['num_cylinders']
    dat['age_mileage_interaction'] = dat['vehicle_age'] * dat['mileage']
    dat['horsepower_age_interaction'] = dat['horsepower'] * dat['vehicle_age']

    # Remove outliers for all numerical variables
    dat = remove_outliers(dat)

    return dat

# Apply the preprocessing to your train and test datasets
train_dat = add_preprocess_data(train_dat)
test_dat = add_preprocess_data(test_dat)
```

#### Stacking Top 3

The stacking method combines top-performing models to capture various patterns in the data, enhancing prediction performance. In this case, we trained and utilized the prediction outcomes of the top three models: Gradient Boosting, XGBoost Regressor, and LightGBM Regressor for a final model.


```python
# Initialize each model with desired hyperparameters
lgbm = LGBMRegressor(learning_rate=0.1,
                     n_estimators=1000,
                     max_depth=10)
gb = GradientBoostingRegressor(learning_rate=0.1,
                               n_estimators=1000,
                               max_depth=10)
xgb = XGBRegressor(learning_rate=0.1,
                   n_estimators=1000,
                   max_depth=10,
                   random_state=42)

# Define the base estimators and the final estimator
estimators = [
    ('lgbm', lgbm),
    ('gb', gb),
    ('xgb', xgb)
]

# Define the stacking model with a meta-model
stacking = StackingRegressor(
    estimators=estimators,
    final_estimator=LinearRegression(),
    cv=5,
    n_jobs=-1
)

# Fit train dataset with Stacking model
stacking.fit(X_train, y_train)

# Predict the target variable 'price'
y_train_pred = stacking.predict(X_train)
y_test_pred = stacking.predict(X_test)

# Print the result
print(f' MAPE train: {mean_absolute_percentage_error(y_train, y_train_pred):.5f}')
```

```python
# Create 'price' column in the test dataset
test_dat['price'] = y_test_pred

# Adjust the format
price_data = test_dat[['vin', 'price']]

# Save the output in the csv file.
price_data.to_csv('Stacking.csv', index = False)
```
